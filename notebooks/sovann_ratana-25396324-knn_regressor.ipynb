```python
# Import packages
import pandas as pd
import numpy as np
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import FunctionTransformer
```

```python
# Load dataset
# Import Dataset 
X_train = pd.read_csv('../data/processed/X_train.csv')
X_val = pd.read_csv('../data/processed/X_val.csv')
X_test = pd.read_csv('../data/processed/X_test.csv')
y_train = pd.read_csv('../data/processed/y_train.csv')
y_val = pd.read_csv('../data/processed/y_val.csv')
y_test = pd.read_csv('../data/processed/y_test.csv')
```

```python
from joblib import load
```

```python
# Load preprocessor
preprocessor= load('../models/preprocessor/preprocessor.pkl')
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Create a pipeline with the preprocessor and the k-NN Regressor model
knn_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),  # Step 1: preprocess features
    ('model', KNeighborsRegressor(n_neighbors=5))  # Step 2: train model
])
```

```python
# Fit the pipeline on the training data
knn_pipeline.fit(X_train, y_train)
```

```python
preds_train = knn_pipeline.predict(X_train)
preds_val = knn_pipeline.predict(X_val)
preds_test = knn_pipeline.predict(X_test)
```

```python
from my_krml_ratana.models.performance import print_regressor_scores
```

```python
print_regressor_scores(y_preds=preds_train, y_actuals=y_train, set_name='Training')
print_regressor_scores(y_preds=preds_val, y_actuals=y_val, set_name='Validating')
```

```python
print_regressor_scores(y_preds=preds_test, y_actuals=y_test, set_name='Testing')
```
